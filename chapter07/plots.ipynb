```python
import matplotlib.pyplot as plt

x_coords = [1, 2, 3, 4]
y_coords = [2, 1, 4, 5]

plt.plot(x_coords, y_coords)
plt.xticks([1, 2, 3, 4], ["1. Q", "2. Q", "3. Q", "4. Q"])
plt.yticks([0, 1, 2, 3, 4, 5], ['$0m', '$1m', '$2m', '$3m', '$4m', '$5m'])
plt.title("Sales in 2024")
plt.xlabel("Quaters")
plt.ylabel("Sales")
plt.show()
```
